

# Merging

- on : keyword use করা  হয় যে কমন কলামের উপর ভিত্তি করে আমরা দুইটা DataFrame কে Merge করব।

- How = inner মনে employee_id কলামে যেগুলো মিলে সেগুলো ADD করবে।মানে যাদের keys সেম।(Default) ভাবে How Inner ই হয়।

- How = কমন কলাম employee_id এর উপর ভিত্তি করে সবগুলো আসবে কমন, অঙ্কমন সব।

- How = left মানে লেফট employee table এর basis এ add করবে। লেফট সাইডের টা রাখবে

- How = right মানে salary table এর basis এ add করবে।


```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries, on = 'employee_id') # default how = inner
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'outer')
```

```python
pd.merge(employees, salaries, on='employee_id', how='left')
```

```python
pd.merge(employees, salaries , on = 'employee_id' , how = 'right')
```

# Conctenation
Concatanation হয়তো vertically নয়ত horizontally add করবে। <br><br>

যদি আমরা পুরো dataframe কে concatenation করতে চাই তাহলে concat function ব্যবহার করব। <br> <br>

আর যদি আমরা dataframe কে শুধু একটা column এর ভিত্তিতে করতে চাই তাহলে merge করব। <br> <br>

আমরা যখন concat করব আর যখন axis দিব না তখন
Default প্রথম dataset পর দ্বিতীয় ডাটাসেট যুক্ত হবে।  <br><br>

আর যখন axis = 1 দিব তখন কলম wise add হবে।<br>

- ignore_index = True দিয়ে দিলে default ভাবে ০ , ১ সেট হবে।

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
pd.concat([df1,df2] , axis = 1)
```

```python
pd.concat([df1,df2], axis = 1 , ignore_index=True)
```

# Pivot Table

- Pivot Table value হিসেবে Quantitive value থাকবে। <br>
- default value হিসেবে means function থাকবে

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values = "Sales",index = 'Region',columns="Product",aggfunc = 'median')
```

```python
pd.pivot_table(df, values = ['Sales' ,'Units'] , index = 'Region', columns = 'Product')
```

# Cross Table

```python
pd.crosstab(df['Product'] , df['Region'])
```
